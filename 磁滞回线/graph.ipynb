```python
import matplotlib.pyplot as plt

# 第一组数据
H1 = [2597.0132, 2227.6989, 1861.7181, 1506.4045, 1166.425, 843.7797, 549.8025, 288.4936, 67.1867, -130.119, 
      -300.7567, -457.3937, -604.6969, -747.3332, -879.969, -1018.6051, -1154.57, -1295.8773, -1449.1808,
      -1607.1512, -1771.1219, -1959.0938, -2161.7331, -2380.3732, -2695.6848]

B1 = [360.4, 353.3, 345.7, 336.5, 325, 310.9, 292.5, 269.2, 239.9, 207, 170.1, 131.1, 90.7, 49.6, 7, -34.7, 
      -76.8, -118.1, -157.6, -196.4, -234.3, -268.6, -300.7, -330.4, -345.6]

# 第二组数据
H2 = [2660.3497, 2407.7079, 2167.0667, 1971.0944, 1775.1221, 1623.152, 1462.5148, 1309.2113, 1170.5752, 1026.6055,
      891.3029, 745.3331, 617.3642, 469.3943, 308.7571, 132.7858, -67.8534, -301.1609, -570.4702, -869.781,
      -1197.0932, -1541.7396, -1897.0532, -2264.3674, -2635.6818]

B2 = [350.9, 326.3, 299.9, 266.8, 233.7, 194, 155.6, 116.1, 74.4, 33.5, -8.7, -49.3, -92.6, -132.9, -171.3,
      -207.4, -239.8, -267.3, -289.4, -307, -320.4, -331.2, -340.4, -347.8, -354.6]

# 创建图表
# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体，需确保系统中有此字体
plt.rcParams['axes.unicode_minus'] = False     # 正常显示负号
plt.figure(figsize=(10, 7))           # 设置图表大小
# 突出显示xy坐标轴
plt.axhline(0, color='black', linewidth=2)   # 横轴
plt.axvline(0, color='black', linewidth=2)   # 纵轴
# 绘制第一组曲线
plt.plot(H1, B1, label='第一组数据', marker='o', linestyle='-', color='b')
for i in range(len(H1)):
    plt.text(H1[i], B1[i], f'({H1[i]:.1f}, {B1[i]:.1f})', fontsize=8, color='blue', ha='left', va='bottom')
# 绘制第二组曲线
plt.plot(H2, B2, label='第二组数据', marker='x', linestyle='-', color='r')
for i in range(len(H2)):
    plt.text(H2[i], B2[i], f'({H2[i]:.1f}, {B2[i]:.1f})', fontsize=8, color='red', ha='right', va='top')

# 查找与X轴和Y轴的交点
def find_intersections(H, B):
    intersections = []
    for i in range(1, len(H)):
        if B[i-1] * B[i] < 0:  # 与X轴交点
            intersections.append((H[i], 0))
        if H[i-1] * H[i] < 0:  # 与Y轴交点
            intersections.append((0, B[i]))
    return intersections

intersections1 = find_intersections(H1, B1)
intersections2 = find_intersections(H2, B2)

# 标注交点
for x, y in intersections1 + intersections2:
    plt.scatter(x, y, color='green')
    plt.text(x, y, f'({x:.1f}, {y:.1f})', fontsize=10, color='green', ha='right', va='top')
# 设置标题和标签
plt.title('模具钢的磁滞回线', fontsize=16)
plt.xlabel('修正后的 H (A/m)', fontsize=14)
plt.ylabel('B (mT)', fontsize=14)

# 添加网格
plt.grid(True)

# 显示图例
plt.legend()

# 展示图表
plt.show()
```
