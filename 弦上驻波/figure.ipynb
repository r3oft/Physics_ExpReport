```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 设置中文字体（如果需要）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 数据 (根据表格中 ln T 与 ln f1 的值)
lnT = np.array([0.910, 1.603, 2.008, 2.296, 2.519])
lnf = np.array([3.360, 3.712, 3.885, 4.027, 4.128])

# 定义线性函数
def linear(x, a, b):
    return a * x + b

# 进行线性拟合
popt, pcov = curve_fit(linear, lnT, lnf)
slope, intercept = popt

# 生成平滑曲线
x_fit = np.linspace(min(lnT), max(lnT), 100)
y_fit = linear(x_fit, slope, intercept)

# 绘图
plt.figure(figsize=(8, 6))
plt.scatter(lnT, lnf, label='实验数据')
plt.plot(x_fit, y_fit, 'r-', label=f'拟合直线: ln f = {slope:.3f} ln T + {intercept:.3f}')

plt.xlabel('ln T')
plt.ylabel('ln f')
plt.title('ln f 与 ln T 的关系')
plt.legend()
plt.grid(True)
plt.show()

print(f'线性拟合结果: slope = {slope:.3f}, intercept = {intercept:.3f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 弦号
strings = ['1', '2', '3', '6', '7', '8', '12']

# 表格中给出的数据
ln_mu = np.array([-5.164, -5.696, -5.263, -5.599, -5.664, -6.840, -5.155])
ln_f1 = np.array([3.712, 3.935, 3.685, 3.912, 3.875, 4.483, 3.786])

# 定义线性函数
def linear_function(x, a, b):
    return a * x + b

# 线性拟合
popt, pcov = curve_fit(linear_function, ln_mu, ln_f1)
slope, intercept = popt

# 生成用于绘制平滑曲线的横坐标
x_fit = np.linspace(min(ln_mu), max(ln_mu), 100)
y_fit = linear_function(x_fit, slope, intercept)

# 绘图
plt.figure(figsize=(8, 6))
plt.scatter(ln_mu, ln_f1, label='实验数据', color='blue')
plt.plot(x_fit, y_fit, 'r-', label=f'拟合直线: ln(f) = {slope:.3f}·ln(μ) + {intercept:.3f}')

# 在图中标注弦号
for i, label in enumerate(strings):
    plt.text(ln_mu[i], ln_f1[i], label, fontsize=9, ha='right', va='bottom')

plt.xlabel('ln(μ)')
plt.ylabel('ln(f₁)')
plt.title('不同线密度琴弦的基频关系')
plt.grid(True)
plt.legend()
plt.show()

print(f'线性拟合结果: slope = {slope:.3f}, intercept = {intercept:.3f}')
```
