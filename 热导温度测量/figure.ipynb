```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 温度 (°C)
t = np.array([30.5, 35.8, 39.8, 44.9, 50.1])

# 对应的电动势 (mV)
E_x = np.array([1.08, 1.34, 1.51, 1.70, 1.89])

# 定义线性函数
def linear_func(x, a, b):
    return a * x + b

# 进行线性拟合
popt, pcov = curve_fit(linear_func, t, E_x)
slope, intercept = popt

# 生成平滑曲线
t_fit = np.linspace(min(t), max(t), 100)
E_fit = linear_func(t_fit, slope, intercept)

# 绘图
plt.figure(figsize=(8, 6))
plt.scatter(t, E_x, color='blue', label='实验数据')
plt.plot(t_fit, E_fit, 'r-', label=f'拟合直线: $E_x$ = {slope:.4f}·$t$ + {intercept:.4f}')

plt.xlabel('温度 $t$ (°C)')
plt.ylabel('热电势 $E_x$ (mV)')
plt.title('E_x-t 温度曲线')
plt.legend()
plt.grid(True)
plt.show()

print(f'线性拟合结果: slope = {slope:.4f} mV/°C, intercept = {intercept:.4f} mV')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 1. 数据定义
# 温度 t(°C)
t = np.array([30.3, 35.4, 40.1, 44.9, 50.1])
# 1/T(K^-1)
inv_T = np.array([0.00330, 0.00324, 0.00319, 0.00314, 0.00309])
# 热敏电阻 R_T(Ω)
R_T = np.array([2243.1, 1811.1, 1503.6, 1243.0, 1013.0])
# ln(R_T)
ln_R_T = np.array([7.7156, 7.5017, 7.3156, 7.1253, 6.9207])

# 2. 绘制 R_T - t 关系曲线
plt.figure(figsize=(8,6))
plt.scatter(t, R_T, color='blue', label='实验数据')
plt.plot(t, R_T, 'r--', alpha=0.5)  # 简单连线，可根据需求去掉或调整
plt.xlabel('温度 t (°C)')
plt.ylabel('热敏电阻 R_T (Ω)')
plt.title('$R_T - t$ 关系曲线')
plt.legend()
plt.grid(True)
plt.show()

# 3. 对 ln(R_T) - 1/T 关系进行线性拟合
def linear_func(x, a, b):
    return a * x + b

popt, pcov = curve_fit(linear_func, inv_T, ln_R_T)
a, b = popt

# 4. 生成平滑曲线
x_fit = np.linspace(min(inv_T), max(inv_T), 100)
y_fit = linear_func(x_fit, a, b)

# 5. 绘制 ln(R_T) - 1/T 关系及线性拟合图
plt.figure(figsize=(8,6))
plt.scatter(inv_T, ln_R_T, color='blue', label='实验数据')
plt.plot(x_fit, y_fit, 'r-', label=f'拟合直线: ln(R_T) = {a:.4f}·(1/T) + {b:.4f}')
plt.xlabel('$1/T$ (K$^{-1}$)')
plt.ylabel('ln($R_T$)')
plt.title('ln($R_T$) - $1/T$ 关系')
plt.legend()
plt.grid(True)
plt.show()

print(f'线性拟合结果: a = {a:.4f}, b = {b:.4f}')
```
