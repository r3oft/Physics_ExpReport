```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.font_manager as fm

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 实验数据
M = np.array([500, 750, 1000, 1250, 1500, 1750, 2000, 2250])
l_avg = np.array([0.750, 1.015, 1.275, 1.525, 1.780, 2.055, 2.315, 2.550])

# 定义线性函数
def linear(x, a, b):
    return a * x + b

# 拟合数据
popt, pcov = curve_fit(linear, M, l_avg)
slope = popt[0]
intercept = popt[1]

# 绘制数据点和拟合直线
plt.scatter(M, l_avg, label='数据点')
plt.plot(M, linear(M, *popt), color='red', label=f'拟合直线: y = {slope:.6f}x + {intercept:.2f}')
plt.xlabel('砝码质量 M (g)')
plt.ylabel('平均值 $\overline{l}$ (mm)')
plt.title('作图法求钼丝杨氏模量')
plt.legend()
plt.grid(True)
plt.show()

# 输出斜率
print(f'拟合直线的斜率: {slope:.9f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 设置中文字体（如果需要）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 实验数据
x_over_L = np.array([0.110, 0.139, 0.167, 0.194, 0.250, 0.278, 0.306, 0.333])
f = np.array([893.200, 892.400, 891.700, 891.500, 891.799, 892.700, 893.591, 894.594])

# 绘制数据点
plt.scatter(x_over_L, f, label='实验数据')

# 定义拟合函数（根据需要选择合适的函数形式）
def quadratic(x, a, b, c):
    return a * x**2 + b * x + c

# 对数据进行二次函数拟合
popt, pcov = curve_fit(quadratic, x_over_L, f)
a, b, c = popt

# 生成拟合曲线
x_fit = np.linspace(min(x_over_L), max(x_over_L), 100)
f_fit = quadratic(x_fit, a, b, c)
plt.plot(x_fit, f_fit, 'r-', label='拟合曲线')

# 设置图形参数
plt.xlabel('悬挂点位置 $x/L$')
plt.ylabel('共振频率 $f$ (Hz)')
plt.title('不锈钢金属棒共振频率与悬挂位置关系图')
plt.legend()
plt.grid(True)
plt.show()

# 输出拟合参数
print(f'拟合二次函数参数: a = {a:.6f}, b = {b:.6f}, c = {c:.6f}')
```
