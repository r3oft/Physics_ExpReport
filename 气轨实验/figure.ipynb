```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 数据
m = np.array([0.21735, 0.24247, 0.25489, 0.27972, 0.29209])
T_squared = np.array([2.4940, 2.7766, 2.9156, 3.1955, 3.3341])

# 定义线性函数
def linear_func(x, a, b):
    return a * x + b

# 拟合
popt, _ = curve_fit(linear_func, m, T_squared)
a, b = popt

# 作图
m_fit = np.linspace(min(m), max(m), 100)
T_fit = linear_func(m_fit, a, b)

plt.figure(figsize=(8, 6))
plt.scatter(m, T_squared, color='blue', label='实验数据')
plt.plot(m_fit, T_fit, 'r--', label=f'拟合: $T^2$ = {a:.4f}·m + {b:.4f}')
plt.xlabel('m (kg)')
plt.ylabel('$T^2$ (s$^2$)')
plt.title('$T^2$ - $m$ 拟合直线图')
plt.legend()
plt.grid(True)
plt.show()

print(f'线性表达式: T^2 = {a:.4f}·m + {b:.4f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 数据
delta_s_cm = np.array([1, 3, 5, 10])  # cm
delta_s = delta_s_cm / 100           # m
delta_t_ms = np.array([29.04, 86.44, 143.01, 279.98])  # ms
delta_t = delta_t_ms / 1000                              # s
v = np.array([0.344, 0.347, 0.349, 0.357])               # m/s

def linear_func(x, a, b):
    return a*x + b

# 1) 在 v-Δs 上做线性拟合（结果会显示斜率接近 0）
popt_s, _ = curve_fit(linear_func, delta_s, v)
a_s, b_s = popt_s

# 2) 在 v-Δt 上做线性拟合（同样斜率很小）
popt_t, _ = curve_fit(linear_func, delta_t, v)
a_t, b_t = popt_t

# 直接求均值作为近似常数速度
v_mean = np.mean(v)

# 绘图1：v-Δs
s_fit = np.linspace(0, max(delta_s)*1.1, 50)
v_fit_s = linear_func(s_fit, a_s, b_s)
plt.figure(figsize=(8,6))
plt.scatter(delta_s, v, c='b', label='测量数据')
plt.plot(s_fit, v_fit_s, 'r--', label=f'线性拟合: y={a_s:.4f}x+{b_s:.4f}')
plt.axhline(v_mean, color='g', linestyle=':', label=f'恒定速度: v={v_mean:.4f} m/s')
plt.xlabel('Δs (m)')
plt.ylabel('v (m/s)')
plt.legend()
plt.title('v - Δs')
plt.grid(True)
plt.show()

# 绘图2：v-Δt
t_fit = np.linspace(0, max(delta_t)*1.1, 50)
v_fit_t = linear_func(t_fit, a_t, b_t)
plt.figure(figsize=(8,6))
plt.scatter(delta_t, v, c='b', label='测量数据')
plt.plot(t_fit, v_fit_t, 'r--', label=f'线性拟合: y={a_t:.4f}x+{b_t:.4f}')
plt.axhline(v_mean, color='g', linestyle=':', label=f'恒定速度: v={v_mean:.4f} m/s')
plt.xlabel('Δt (s)')
plt.ylabel('$v$ (m/s)')
plt.legend()
plt.title('v - Δt')
plt.grid(True)
plt.show()

print(f'v-Δs 拟合斜率: {a_s:.4e}, 截距: {b_s:.4f}')
print(f'v-Δt 拟合斜率: {a_t:.4e}, 截距: {b_t:.4f}')
print(f'简单取平均值, v_mean = {v_mean:.4f} m/s')
```
