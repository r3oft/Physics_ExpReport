```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 设置中文字体（如果需要）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 实验数据
I_S = np.array([0, 0.50, 1.00, 1.50, 2.00, 2.50, 3.00])            # 工作电流 I_S (mA)
V_H = np.array([0.0, 27.2, 54.2, 81.3, 108.6, 135.3, 162.4])       # 霍尔电压 V_H (mV)

# 绘制实验数据点
plt.scatter(I_S, V_H, label='实验数据')

# 定义线性函数用于拟合
def linear(x, k, b):
    return k * x + b

# 进行线性拟合
popt, pcov = curve_fit(linear, I_S, V_H)
k, b = popt

# 生成拟合曲线数据
I_S_fit = np.linspace(min(I_S), max(I_S), 100)
V_H_fit = linear(I_S_fit, k, b)

# 绘制拟合直线
plt.plot(I_S_fit, V_H_fit, 'r-', label=f'拟合直线: $V_H$ = {k:.2f}$I_S$ + {b:.2f}')

# 图形设置
plt.xlabel('工作电流 $I_S$ (mA)')
plt.ylabel('霍尔电压 $V_H$ (mV)')
plt.title('$V_H$ 与 $I_S$ 的关系曲线')
plt.legend()
plt.grid(True)

# 显示图形
plt.show()

# 输出拟合结果
print(f'拟合得到的斜率 k = {k:.2f} mV/mA，截距 b = {b:.2f} mV')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 设置中文字体（如果需要）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 实验数据

# B 与 I_M 的关系数据
I_M = np.array([0, 50, 100, 150, 200, 250, 300])        # 励磁电流 I_M (mA)
B = np.array([0.0, 38.3, 75.2, 113.2, 150.6, 188.2, 225.1])   # 磁感应强度 B (mT)

# V_H 与 B 的关系数据
V_H = np.array([0.0, 13.8, 27.2, 41.0, 54.6, 68.2, 81.6])     # 霍尔电压 V_H (mV)

# 绘制 B 与 I_M 的关系图
plt.figure()
plt.scatter(I_M, B, label='实验数据')

# 定义线性函数用于拟合
def linear_func(x, k, b):
    return k * x + b

# 对 B 与 I_M 数据进行线性拟合
popt, pcov = curve_fit(linear_func, I_M, B)
k_B_IM, b_B_IM = popt

# 生成拟合曲线数据
I_M_fit = np.linspace(min(I_M), max(I_M), 100)
B_fit = linear_func(I_M_fit, k_B_IM, b_B_IM)

# 绘制拟合直线
plt.plot(I_M_fit, B_fit, 'r-', label=f'拟合直线: B = {k_B_IM:.2f} × $I_M$ + {b_B_IM:.2f}')
plt.xlabel('励磁电流 $I_M$ (mA)')
plt.ylabel('磁感应强度 $B$ (mT)')
plt.title('$B$ 与 $I_M$ 的关系图')
plt.legend()
plt.grid(True)
plt.show()

# 输出拟合结果
print(f'B 对 I_M 的线性拟合结果: 斜率 k = {k_B_IM:.4f} mT/mA，截距 b = {b_B_IM:.4f} mT')

# 绘制 V_H 与 B 的关系图
plt.figure()
plt.scatter(B, V_H, label='实验数据')

# 对 V_H 与 B 数据进行线性拟合
popt, pcov = curve_fit(linear_func, B, V_H)
k_VH_B, b_VH_B = popt

# 生成拟合曲线数据
B_fit_VH = np.linspace(min(B), max(B), 100)
V_H_fit = linear_func(B_fit_VH, k_VH_B, b_VH_B)

# 绘制拟合直线
plt.plot(B_fit_VH, V_H_fit, 'r-', label=f'拟合直线: $V_H$ = {k_VH_B:.2f} × B + {b_VH_B:.2f}')
plt.xlabel('磁感应强度 $B$ (mT)')
plt.ylabel('霍尔电压 $V_H$ (mV)')
plt.title('$V_H$ 与 $B$ 的关系图')
plt.legend()
plt.grid(True)
plt.show()

# 输出拟合结果
print(f'V_H 对 B 的线性拟合结果: 斜率 k = {k_VH_B:.4f} mV/mT，截距 b = {b_VH_B:.4f} mV')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# 轴向距离 X (mm)
X = np.array([-25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25])

# 测量值 B (mT)
B_measured = np.array([0.132, 0.135, 0.139, 0.141, 0.144, 0.145, 0.144, 0.142, 0.141, 0.137, 0.134])

# 计算值 B (mT)
B_calculated = np.array([0.132, 0.136, 0.139, 0.142, 0.143, 0.144, 0.143, 0.142, 0.139, 0.136, 0.132])

# 创建插值函数
f_measured = interp1d(X, B_measured, kind='cubic')
f_calculated = interp1d(X, B_calculated, kind='cubic')

# 生成平滑的 X 轴数据
X_new = np.linspace(-25, 25, 200)

# 计算插值后的 B 值
B_measured_smooth = f_measured(X_new)
B_calculated_smooth = f_calculated(X_new)

# 绘制平滑曲线
plt.figure(figsize=(8, 6))
plt.plot(X_new, B_measured_smooth, label='测量值 B', linewidth=2)
plt.plot(X_new, B_calculated_smooth, label='计算值 B', linewidth=2)

# 原始数据点
plt.scatter(X, B_measured, color='blue', marker='o')
plt.scatter(X, B_calculated, color='orange', marker='s')

# 设置图表
plt.xlabel('轴向距离 X (mm)')
plt.ylabel('磁感应强度 B (mT)')
plt.title('圆电流线圈轴线上磁感应强度分布')
plt.legend()
plt.grid(True)
plt.show()
```
